```python
# Resume Analyzer Function
def analyze_resume(resume):
    score = 0
    feedback = []

    resume = resume.lower()

    # Sections check
    if "name" in resume:
        score += 1
    else:
        feedback.append("Add your name")

    if "skills" in resume:
        score += 2
    else:
        feedback.append("Add skills section")

    if "education" in resume:
        score += 2
    else:
        feedback.append("Add education details")

    if "experience" in resume:
        score += 2
    else:
        feedback.append("Add experience")

    if "project" in resume:
        score += 2
    else:
        feedback.append("Add projects")

    # Email check
    if "@" in resume:
        score += 1
    else:
        feedback.append("Add email address")

    return score, feedback


# Final Suggestion Function
def final_suggestion(score):
    if score >= 8:
        return "Excellent Resume 🔥"
    elif score >= 5:
        return "Good Resume 👍 Improve a few areas"
    else:
        return "Needs Improvement ⚠️ Work on basics"


# Hackathon Requirement (OpenEnv)
class ResumeEnv:
    def __init__(self):
        self.resume = ""

    def reset(self, resume_text):
        self.resume = resume_text
        return self.resume

    def step(self):
        score, feedback = analyze_resume(self.resume)
        reward = score

        return {
            "score": score,
            "feedback": feedback,
            "reward": reward,
            "suggestion": final_suggestion(score)
        }


# Run Program
env = ResumeEnv()

resume = input("Paste your resume here: ")
env.reset(resume)

result = env.step()

print("\n📊 Resume Score:", result["score"], "/ 10")

print("\n📌 Feedback:")

# UPDATED PART (no feedback case handle)
if len(result["feedback"]) == 0:
    print("✅ No issues found. Your resume looks great!")
else:
    for f in result["feedback"]:
        print("-", f)

print("\n💡 Final Suggestion:", result["suggestion"])
```
