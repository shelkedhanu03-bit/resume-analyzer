```python
!pip install -q gradio pdfplumber

import gradio as gr
import pdfplumber

def analyze_resume(file):
    text = ""
    with pdfplumber.open(file.name) as pdf:
        for page in pdf.pages:
            if page.extract_text():
                text += page.extract_text()

    resume = text.lower()

    skills_list = [
        "python","java","c++","sql","excel","machine learning","deep learning",
        "tensorflow","pytorch","opencv","flask","node.js","html","css","javascript",
        "react","git","github","api","data analysis","nlp","computer vision",
        "communication","leadership","management","sales","marketing","finance",
        "accounting","teamwork","problem solving","customer service"
    ]

    found_skills = [skill for skill in skills_list if skill in resume]

    if "intern" in resume or "fresher" in resume:
        level = "Fresher"
    elif "experience" in resume or "worked" in resume or "years" in resume:
        level = "Experienced"
    else:
        level = "Intermediate"

    score = 35 + len(found_skills) * 2

    if "project" in resume:
        score += 10
    if "education" in resume:
        score += 5
    if "summary" in resume or "objective" in resume:
        score += 5
    if "experience" in resume:
        score += 10
    if "achievement" in resume:
        score += 5
    if "certification" in resume:
        score += 5

    score = min(score, 90)

    suggestions = []

    if len(found_skills) < 5:
        suggestions.append("Add more relevant skills")

    if "project" not in resume:
        suggestions.append("Add project or work experience details")

    if "achievement" not in resume:
        suggestions.append("Add achievements or measurable results")

    if "certification" not in resume:
        suggestions.append("Add certifications")

    if "summary" not in resume and "objective" not in resume:
        suggestions.append("Add professional summary")

    if "machine learning" in resume or "data" in resume:
        role = "Data Analyst / ML Engineer"
    elif "sales" in resume or "marketing" in resume:
        role = "Sales / Marketing Executive"
    elif "account" in resume or "finance" in resume:
        role = "Finance / Accounting Role"
    elif "management" in resume:
        role = "Management Role"
    else:
        role = "General Professional Role"

    result = f"""
CAREERPILOT AI REPORT

Experience Level: {level}
ATS Score: {score}/100

Target Role: {role}

Skills Detected:
{", ".join(found_skills)}

Suggestions:
{", ".join(suggestions)}
"""

    return result


iface = gr.Interface(
    fn=analyze_resume,
    inputs=gr.File(label="Upload Any Resume (PDF)"),
    outputs="text",
    title="CareerPilot AI - Universal Resume Analyzer",
    description="Analyze any resume (Tech or Non-Tech) and get ATS score + suggestions"
)

iface.launch()
```
